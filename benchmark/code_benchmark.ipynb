Benchmark of A simple ECC certificate library
=============================================

_Author: Tony Cheneau_

This page provides some benchmark of the simple ECC certificate library. The data has been obtained using a Redbee Econotag.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpltools import style
from mpltools import layout

style.use('ggplot')
```

These three files are a dump of the serial output of the Econotag. The dump was realized using [serialreadts](https://github.com/tcheneau/serialreadts), a tool that prepend a (local) timestamp in front of every lines of a serial output. The Econotag have been loaded with the images of the udp-client.c code with varying curve parameter (hence the file name). The image performs crypto operation in a loop (see the *test_crypto()* function).

```python
log_192k1 = open("speed-secp192k1.txt").read()
log_160k1 = open("speed-secp160k1.txt").read()
log_128r1 = open("speed-secp128r1.txt").read()
```

```python
def parse_lines(data):
    lines = data.splitlines()
    lines = [line.split("]") for line in lines]
    return [(float(a[1:]), b) for [a,b] in lines]    

def extract_timing_text(lines):
    timing_text = [text for (time, text) in lines if "done" in text]
    timing_text += [ text[7:] for text in timing_text]
    return timing_text


def parse_log(data, curve_name):
    lines = parse_lines(data)
    timing_text = extract_timing_text(lines)
    start = []
    stop = []
    i = 0
    for (time, text) in lines:
        if text not in timing_text:
            continue
        if i%2 == 0 and 'done' in text:
            continue

        if i%2 == 0:
            start.append((time, text))
        else:
            stop.append((time,text))
        i += 1
        
    timings = []
    length = min(len(start), len(stop))
    for i in range(length):
        (start_time, text) = start[i]
        (stop_time, _) = stop[i]
        timings.append((curve_name, text, stop_time - start_time))
        
    return timings
    
```

```python
timings = parse_log(log_128r1, "secp128r1")
timings += parse_log(log_160k1, "secp160k1")
timings += parse_log(log_192k1, "secp192k1")
```

Parse the data and make sure it looks OK.

```python
df = pd.DataFrame(timings, columns = ["curve", "event", "duration"])
df = df.sort("curve")
df.head()
```

```python
# here we remove uninteresting "event"
mask = (df['event']!="ECDH derive key") & \
   (df['event']!="verifying ECDSA signature (bad 1/2)") & \
   (df['event']!="verifying ECDSA signature (bad 2/2)") & \
   (df['event']!="verify a client certificate") & \
   (df['event']!="deserialization") & \
   (df['event']!="serialization")
df = df[mask]
```

```python
def rename_event(event_name):
    if event_name == "sign random data":
        return "ECDSA sign 200 bytes"
    elif event_name == "verify own's certificate":
        return "ECDSA verify 200 bytes"
    elif event_name == "generate a random certificate":
        return "Generate certificate"
    elif event_name == "verifying ECDSA signature (good)":
        return "Verify a certificate"
    elif event_name == "ECDH first message":
        return "ECDH first event"
    elif event_name == "ECDH second message":
        return "ECDH second event"
    else:
        return event_name
    

df['event'] = df['event'].apply(rename_event)
```

```python
# count the number of events we have in our dataset
# this is to make sure we have enough "raw" material
df.groupby(['curve','event']).apply(lambda x: len(x))
```

The results below are in seconds.

```python
res = df.groupby(["curve", "event"]).agg(np.mean)
res
```

```python
# same data, presented differently
res.unstack(0)
```

```python
unstacked_res =  res.unstack(0)
unstacked_res.columns = [curve for (name, curve) in  unstacked_res.columns]

del unstacked_res.index.name
#unstacked_res.sort_index(axis=1)
#unstacked_res.index.name = "Cryptographic primitive"
unstacked_res.plot(kind="bar")
plt.xticks(rotation=65)
plt.ylabel("duration (in seconds)")
plt.legend(title="Curves", loc=2)
plt.show()
# plt.savefig("crypto-speed.pdf", bbox_inches="tight", dpi=300)
```
